```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def comb(n, r):
    return np.math.factorial(n) / (np.math.factorial(r)*np.math.factorial(n-r))
```

## Function for the probability of a correct decision

```python
def p_correct_decision(n, p):
    h = n//2 + 1
    p_correct_decision = 0
    
    for m in range(h, n+1):
        p_correct_decision += comb(n,m) * p**m * (1-p)**(n-m)
    
    return p_correct_decision
```

```python
# Test if p_correct_decision function is correct with 1a
n = 39; p = 0.6 
print(f"P(correct_decision) = {p_correct_decision(n,p)}")
```

## Plotting

```python
jury_sizes = np.arange(4,10 + 1, step = 3) 
print(f"Jury sizes = {jury_sizes}")
competence_levels = np.arange(0.1,1 + 0.1, step=0.4) 
print(f"Competence levels = {competence_levels}")

plt.figure(figsize=(15,5))

bar_labels = []

# For each combination of jury size and competence level plot a line
i = 0
for p in competence_levels:
    for n in jury_sizes:
        bar_label = f"n={n}, p={p}"
        plt.bar(i, p_correct_decision(n,p), label=bar_label, zorder=2)
        bar_labels.append(bar_label)
        i += 1 

plt.xlabel("Parameters")
plt.xticks(ticks=np.arange(len(jury_sizes)*len(competence_levels)), labels=bar_labels)
plt.ylabel("P(correct decision)")
plt.title("P(correct decision) for various parameters")
plt.legend() 
plt.grid(zorder=1) 
plt.show() 
```
