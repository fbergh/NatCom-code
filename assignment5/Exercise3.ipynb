```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def comb(n, r):
    return np.math.factorial(n) / (np.math.factorial(r)*np.math.factorial(n-r))
```

## Question (a)

```python
total = 0
# First part: without strong classifier
for m in range(6,11):
    total += comb(10,m) * 0.6**m * 0.4**(10-m) * 0.15
# Second part: with strong classifier
for m in range(5,11):
    total += comb(10,m) * 0.6**m * 0.4**(10-m) * 0.85

print("P_max", total)
```

## Question (b)

```python
def weighted_function(w):
    total = 0 
    # First part: without strong classifier
    low = 5+int(w/2)+1
    for m in range(low,11):
        total += comb(10,m) * 0.6**m * 0.4**(10-m) * 0.15
    # Second part: with strong classifier
    low = 4-int(w/2)+1
    for m in range(low,11):
        total += comb(10,m) * 0.6**m * 0.4**(10-m) * 0.85
    return total
```

```python
ws = np.arange(1, 11, 1).astype(np.int16)
ps = np.array([weighted_function(w) for w in ws])
max_ps = np.argwhere(ps == np.amax(ps))
plt.figure(figsize=(10,8))
#plt.title('$P_{max}$ for different values of $w$', fontsize=22)
plt.xlabel('$w$', fontsize=22)
plt.ylabel('$P_{max}$', fontsize=22)
plt.plot(ws, ps, linewidth=3)
for p in max_ps:
    plt.axvline(x=ws[p], color='orange',linewidth=3, linestyle=':')
plt.xticks(ws)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
plt.grid()
plt.savefig('ex3b.pdf')
plt.show()
```
